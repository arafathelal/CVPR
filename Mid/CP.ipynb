```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)

class NeuralNetwork(object):
    def __init__(self):
        inputLayerNeurons = 2
        hidden1Neurons = 10
        hidden2Neurons = 6
        outLayerNeurons = 1

        self.learning_rate = 0.2
        self.W_HI  = np.random.randn(inputLayerNeurons, hidden1Neurons)
        self.W_H12 = np.random.randn(hidden1Neurons, hidden2Neurons)
        self.W_OH  = np.random.randn(hidden2Neurons, outLayerNeurons)

    def sigmoid(self, x, der=False):
        if der:
            return x * (1 - x)
        return 1 / (1 + np.exp(-x))

    def feedForward(self, X):
        hidden1_input = np.dot(X, self.W_HI)
        self.hidden1_output = self.sigmoid(hidden1_input)
        hidden2_input = np.dot(self.hidden1_output, self.W_H12)
        self.hidden2_output = self.sigmoid(hidden2_input)
        output_input = np.dot(self.hidden2_output, self.W_OH)
        pred = self.sigmoid(output_input)
        return pred
```
